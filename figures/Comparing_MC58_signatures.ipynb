```python
data_with_groups = data.copy()
```

```python
data_with_groups['group'] = data_with_groups.index.str[0:3]
data_with_groups
```

```python
data_with_group = data.copy()
data_with_group['group'] = data_with_group.index.str[0]
data_with_group
```

```python
colors = ["#c2e699", "#78c679", "#238443", "#fed98e", "#fe9929", "#cc4c02", "#f7f7f7"]
colors2 = ["#238443", "#cc4c02", "#f7f7f7"]
```

```python
network_pal=sns.color_palette(colors, len(data_with_groups.group.unique()))
network_lut = dict(zip(data_with_groups.group.unique(), network_pal))
networks = data_with_groups.group
network_colors = pd.Series(networks).map(network_lut)

network_pal=sns.color_palette(colors2, len(data_with_group.group.unique()))
network_lut = dict(zip(data_with_group.group.unique(), network_pal))
networks = data_with_group.group
network_colors2 = pd.Series(networks).map(network_lut)
```

```python
cmap = sns.diverging_palette(h_neg=210, h_pos=350, s=90, l=30, as_cmap=True)
o = sns.clustermap(data_with_groups.iloc[:,:-1],
                   vmax= .1,
                   linewidths=.005,
                   row_colors=network_colors,
                   col_colors=network_colors2,
                   cmap=cmap)
o.ax_heatmap.set_xticklabels('')
o.ax_heatmap.set_yticklabels('')
#o.fig.suptitle("")
o.savefig("MC58_signature_comparison.png")
p = sns.set_context("talk")
```
