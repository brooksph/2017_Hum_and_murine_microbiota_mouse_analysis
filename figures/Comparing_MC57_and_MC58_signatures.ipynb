```python
data_with_groups = data.copy()
```

```python
data_with_groups['group'] = data_with_groups.index.str[0:4]
data_with_groups
```

```python
data_with_group = data.copy()
data_with_group['Microbiota'] = data_with_group.index.str[1:2]
data_with_groups
```

```python
#['CCI1', 'CCI2', 'CCUT', 'CHI1', 'CHI2', 'CHUT', 'INOC', 'NCI1','NCI2', 'NCUT', 'NHI1', 'NHI2', 'NHUT']
#Conventional C57 = Shades of Orange 
#Conventional NOD = Shades of red 
#Humanized C57 = Shades of Green 
#Humanized NOD = Shades of blue 
#Inoculum = White 
colors = ['#ffffcc','#ffeda0','#fed976','#c7e9c0','#a1d99b','#74c476','#f7f7f7','#fcbba1','#fc9272','#fb6a4a','#c6dbef','#9ecae1','#6baed6']

#Conventional microbiota is red and humanized micrbiota blue 
colors2 = ['#fb6a4a','#6baed6', "#f7f7f7"]
```

```python
network_pal=sns.color_palette(colors)
network_lut = dict(zip(data_with_groups.group.unique(), network_pal))
networks = data_with_groups.group
network_colors = pd.Series(networks).map(network_lut)

network_pal=sns.color_palette(colors2)
network_lut = dict(zip(data_with_group.Microbiota.unique(), network_pal))
networks = data_with_group.Microbiota
network_colors2 = pd.Series(networks).map(network_lut)
```

```python
import matplotlib.pyplot as plt
```

```python
cmap = sns.diverging_palette(h_neg=210, h_pos=350, s=90, l=30, as_cmap=True)
o = sns.clustermap(data_with_groups.iloc[:,:-1],
                   vmax= .1,
                   linewidths=.005,
                   row_colors=network_colors,
                   col_colors=network_colors2,
                   cmap=cmap)
o.ax_heatmap.set_xticklabels('')
o.ax_heatmap.set_yticklabels('')
sns.set_context("talk")
o.savefig("MC58_and_MC57_signature_comparison.png")
```
