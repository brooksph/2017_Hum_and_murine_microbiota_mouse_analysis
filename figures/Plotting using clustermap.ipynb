```python
from sourmash_lib import fig
import pandas as pd
```

```python
D, labels = fig.load_matrix_and_labels("rhizosphere_scaled100_k21")
```

```python
labels[0]
```

```python
labeltext = [l.split('_')[0] for l in labels]
```

```python
labeltext
```

```python
data = pd.DataFrame(D, index=labeltext, columns=labeltext)
```

```python
data
```

```python
%matplotlib inline
```

```python
import seaborn as sns
```

```python
cmap = sns.diverging_palette(h_neg=210, h_pos=350, s=90, l=30, as_cmap=True)
sns.clustermap(data, col_cluster=True, row_cluster=True, linewidths=.5, figsize=(10, 10), cmap=cmap)
```

```python
data_with_groups = data.copy()
```

```python
data_with_groups['group'] = data_with_groups.index.str[3:6]
data_with_groups
```

```python
network_pal=sns.color_palette('Spectral', len(data_with_groups.group.unique()))
network_lut = dict(zip(data_with_groups.group.unique(), network_pal))
networks = data_with_groups.group
network_colors = pd.Series(networks).map(network_lut)

network_pal=sns.color_palette('Spectral', len(data_with_groups.group.unique()))
network_lut = dict(zip(data_with_groups.group.unique(), network_pal))
networks = data_with_groups.group
network_colors2 = pd.Series(networks).map(network_lut)
```

```python
o = sns.clustermap(data_with_groups.iloc[:,:-1],
                   row_colors=network_colors,
                   col_colors=network_colors2,
                   cmap="YlGnBu")
o.ax_heatmap.set_xticklabels('')
o.ax_heatmap.set_yticklabels('')
```
